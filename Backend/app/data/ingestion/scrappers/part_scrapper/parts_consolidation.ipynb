```python
import pandas as pd
import os
from glob import glob

#print current directoryx
print(os.getcwd())

#list all folders in the current directory
folders = [f for f in os.listdir('.') if os.path.isdir(f)]
print(folders)
```

```python
#consilidating categories data
a90_supra_folder = "a90shop_supra_categories"

#list all folders in the a90shop_supra_categories folder
a90_supra_categories_folders = [folders for folders in os.listdir(a90_supra_folder) if os.path.isdir(os.path.join(a90_supra_folder, folders))]
print(a90_supra_categories_folders)
print(len(a90_supra_categories_folders))
```

```python
#empty dataframe to consolidate all categories data
categories_df = []

#loop through each category folder and read the csv file, then append it to the categories_df
for parts_category in os.listdir(a90_supra_folder):
    file_paths = os.path.join(a90_supra_folder, parts_category, "*.csv")
    for file in glob(file_paths):
        df = pd.read_csv(file)
        categories_df.append(df)

        final_categories_df = pd.concat(categories_df, ignore_index=True)
        print(final_categories_df)
        print(final_categories_df.shape)
```
